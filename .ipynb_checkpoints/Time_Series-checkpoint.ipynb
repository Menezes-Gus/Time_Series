```python
model_ar3 = ARIMA(df['market_value'], order=(3,0,0))
results_ar3 = model_ar3.fit()
results_ar3.summary()
```

```python
model_ar4 = ARIMA(df['market_value'], order=(4,0,0))
results_ar4 = model_ar4.fit()
results_ar4.summary()
```

```python
from scipy.stats.distributions import chi2
```

```python
def LLR_test(mod_1, mod_2, DF=1):
    L1 = mod_1.fit().llf
    L2 = mod_2.fit().llf
    LR = 2*(L2-L1)
    p = chi2.sf(LR, DF).round(6)
    return p
```

```python
LLR_test(model_ar2, model_ar3)
```

```python
LLR_test(model_ar3, model_ar4)
```

```python
model_ar5 = ARIMA(df['market_value'], order=(5,0,0))
results_ar5 = model_ar5.fit()
print(results_ar5.summary())
print(LLR_test(model_ar4, model_ar5))
# como o teste de Log-Likelihood obteve um p-value acima de 0,05% assumimos que o modelo ar5 tem menor poder explicativo que ar4
```
